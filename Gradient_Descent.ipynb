```python
import torch
import matplotlib.pyplot as plt

def function(x):
  return x**4 - 3*x**3 + 2

def gradient(x):
  return 4*x**3 - 9*x**2

x = torch.tensor([5.0], requires_grad=True)

learning_rate = 0.1
iterations = 10

x_values = [x.item()]  # Stores x values for graph
y_values = [function(x).item()]

for i in range(iterations):
  loss = function(x)
  loss.backward()
  with torch.no_grad():
    x -= learning_rate * x.grad
  x_values.append(x.item())
  y_values.append(function(x).item())
  x.grad.zero_()

plt.plot(x_values, y_values, marker='o', linestyle='-')
plt.xlabel("x value")
plt.ylabel("Function Value (y=x^2)")
plt.title("Gradient Descent on y = x^2")
plt.show()

```
